```python
# replace the values ['Iris-setosa','Iris-virginica','Iris-versicolor']  with [1,2,3]

df['species']=df.Species.replace(['Iris-setosa','Iris-virginica','Iris-versicolor'],[1,2,3])
```

```python
df.tail()
```

```python
dfx=df.drop(['Species'],axis="columns")
```

```python
dfx.head()
```

```python
y=df.species
```

```python
x=dfx.drop('species',axis='columns')
```

```python
x
```

### Training Procedure

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=3,train_size=0.7)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

```python
linearmodel=LinearRegression()
logisticmodel=LogisticRegression()
```

```python
# linear model training 

linearmodel.fit(x_train,y_train)
```

```python
y_predicted=linearmodel.predict(x_test)
print(y_predicted)
score=linearmodel.score(x_test,y_test)

print('score of linear model ', score)
```

```python
# logistic model training

logisticmodel.fit(x_train,y_train)
y_pred=logisticmodel.predict(x_test)
print(y_pred)
score=logisticmodel.score(x_test,y_test)
print('score of logistic model',score)
```
